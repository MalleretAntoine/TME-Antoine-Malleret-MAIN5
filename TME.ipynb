```python
#Signer le challence pour se loger 

import subprocess
import tempfile
import os

def sign_challenge_hex(challenge: str, sk_path: str = "sk") -> str:
    """
    Signe le challenge avec la clé privée EC (sk_path),
    retourne la signature DER encodée en hexadécimal (str).
    """
    with tempfile.NamedTemporaryFile(mode="w", encoding="utf-8", delete=False) as tmp_challenge:
        tmp_challenge.write(challenge)
        challenge_path = tmp_challenge.name

    sig_path = None
    try:
        with tempfile.NamedTemporaryFile(delete=False) as tmp_sig:
            sig_path = tmp_sig.name

        # Signer avec OpenSSL (SHA256, DER output)
        subprocess.run(
            ["openssl", "dgst", "-sha256", "-sign", sk_path, "-out", sig_path, challenge_path],
            check=True
        )

        # Lire la signature binaire et encoder en hex
        with open(sig_path, "rb") as f:
            sig_bytes = f.read()
            sig_hex = sig_bytes.hex()

        return sig_hex

    finally:
        os.remove(challenge_path)
        if sig_path:
            os.remove(sig_path)

challenge = "fovea dirts expat prawn faker"
signature = sign_challenge_hex(challenge, sk_path="sk")
print("Signature hex:", signature)
```
